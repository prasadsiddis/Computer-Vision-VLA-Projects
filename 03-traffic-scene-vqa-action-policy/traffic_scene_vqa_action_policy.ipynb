# Traffic Scene VQA and Action Policy

This notebook prototypes a simple Vision-Language-Action flow: answer questions about a driving scene and recommend an action.

```python
scenes = {
    "scene_a": {
        "traffic_light": "red",
        "pedestrians": 1,
        "vehicles_ahead": 2,
        "lane_open": True,
    },
    "scene_b": {
        "traffic_light": "green",
        "pedestrians": 0,
        "vehicles_ahead": 0,
        "lane_open": True,
    },
    "scene_c": {
        "traffic_light": "green",
        "pedestrians": 0,
        "vehicles_ahead": 1,
        "lane_open": False,
    },
}

questions = [
    "Is it safe to move?",
    "Are there pedestrians ahead?",
    "What should the car do at the traffic light?",
]
```

```python
def answer_question(scene, question):
    q = question.lower()

    if "pedestrian" in q:
        if scene["pedestrians"] > 0:
            return "Yes. A pedestrian is present ahead.", "yield"
        return "No pedestrians are detected ahead.", "proceed_if_clear"

    if "traffic light" in q:
        light = scene["traffic_light"]
        if light == "red":
            return "The traffic light is red.", "stop"
        if light == "yellow":
            return "The traffic light is yellow.", "slow_down"
        return "The traffic light is green.", "proceed_if_clear"

    if "safe" in q or "move" in q:
        if scene["traffic_light"] == "red":
            return "No. The red light requires stopping.", "stop"
        if scene["pedestrians"] > 0:
            return "No. A pedestrian is in the scene.", "yield"
        if not scene["lane_open"]:
            return "Not yet. The lane is blocked or uncertain.", "slow_down"
        return "Yes. The lane is open and the light permits movement.", "proceed"

    return "The question is outside this starter policy.", "hold"


for scene_id, scene in scenes.items():
    print("\n", scene_id, scene)
    for question in questions:
        answer, action = answer_question(scene, question)
        print("Q:", question)
        print("A:", answer)
        print("Action:", action)
```
